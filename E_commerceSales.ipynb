```python
import pandas as pd #Data Cleaning
import plotly.express as px #Viz
import plotly.graph_objects as go  # Helps giving coustomized Graph
import plotly.io as pio #Templates
import plotly.colors as colors
import numpy as np
pio.templates.default = "plotly_white"
```

```python
path = ("/content/sample_data/ecommerce.csv")
df = pd.read_csv(path, encoding='latin-1') # Try 'latin-1' first, Latin-1 helps machine to read special characters
# If 'latin-1' doesn't work, try other encodings like 'ISO-8859-1', 'cp1252', etc.
# df = pd.read_csv(path, encoding='ISO-8859-1')
# df = pd.read_csv(path, encoding='cp1252')
df.head(5)
```

```python
df.describe()   #Doesnot work on string data, here for example the country name doesnt come to a picture
```

```python
df.info()
```

Date shows object means string ype of data, first thing is to convert that object type to date time .


Converting DATE columns


```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df.info().  # Here you will see that order date and ship date(data type) has been changed to Objecttime64
```

```python
df.head()
```

```python
df['Order Month'] = df['Order Date'].dt.month
df['Order Year'] = df['Order Date'].dt.year
df['Order Day of Week'] = df['Order Date'].dt.dayofweek
```

```python
df.head()
```

#Monthly Sales Analysis

```python
sales_by_month = df.groupby('Order Month')['Sales'].sum().reset_index()
fig = px.line(sales_by_month, x='Order Month', y='Sales', title='Monthly Sales Analysis')
fig.show()
```

```python
df.head()
```

#Sales by Category  

```python
sales_by_category = df.groupby('Category')['Sales'].sum().reset_index()
fig = px.pie(sales_by_category, values='Sales', names='Category', title='Sales by Category')
fig.show()
```

# Sales Analysis by Sub Category

```python
sales_by_subcategory = df.groupby('Sub-Category')['Sales'].sum().reset_index()
fig = px.bar(sales_by_subcategory, x='Sub-Category', y='Sales', title='Sales Analysis by Sub-Category')
fig.show()
```

#Monthly Profit Analysis

```python
df.head()
```

```python
profit_by_month = df.groupby('Order Month')['Profit'].sum().reset_index()
fig = px.line(profit_by_month, x='Order Month', y='Profit', title='Monthly Profit Analysis')
fig.show()
```

# Profit by Category


```python
profit_by_category = df.groupby('Category')['Profit'].sum().reset_index()
fig = px.bar(profit_by_category, x='Profit', y='Category', title='Profit by Category')
fig.show()
```

#Profit by Sub-Category

```python
profit_by_subcategory = df.groupby('Sub-Category')['Profit'].sum().reset_index()
fig = px.bar(profit_by_subcategory, x='Profit', y='Sub-Category', title='Profit by Sub-Category')
fig.show()
```

```python
df.head()
```

#Sale and profit - Customer Segment

```python
sales_profit_by_segment = df.groupby('Segment').agg({'Sales': 'sum', 'Profit': 'sum'})
fig = px.bar(sales_profit_by_segment, x=sales_profit_by_segment.index, y=['Sales', 'Profit'],
             barmode='group', title='Sales and Profit by Customer Segment')
fig.show()
```

#Sales to Profit Ratio

```python
sales_profit_by_segment['Sales to Profit Ratio'] = sales_profit_by_segment['Sales'] / sales_profit_by_segment['Profit']
print(sales_profit_by_segment)
```
